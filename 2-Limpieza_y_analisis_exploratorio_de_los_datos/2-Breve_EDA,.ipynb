```python
# Importamos las librerias necesarias
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Cargamos el dataset
df_burguer = pd.read_csv("dataset/burger_simplify.csv")
df_burguer.head()
```

Veamos los estaditicos basicos del dataset

```python
df_burguer.describe()
```

Ahora veamos un grafico de dispersion de las variables del dataset y posteriormente midamos su correlacion

```python
# Creamos un scatterplot para visualizar la distribución de los datos
sns.set_style("whitegrid")
plt.figure(figsize=(10, 8))
sns.scatterplot(x='Calories', y='Cholesterol (mg)', data=df_burguer)
plt.title('Scatterplot de Calorías vs. Colesterol (mg)')
plt.xlabel('Calorías')
plt.ylabel('Colesterol (mg)')
plt.show()

# Calculamos el coeficiente de correlacion de Pearson
correlation = df_burguer['Calories'].corr(df_burguer['Cholesterol (mg)'])
print(f"Coeficiente de correlación de Pearson entre Calorías y Colesterol: {correlation:.2f}")
```

El gráfico no es "malo" en absoluto. De hecho, concidero que puede ser un buen punto de partida. La visualizacion de los datos es clara y nos da una idea de cuales pudieran ser algunos clusters futuros.
